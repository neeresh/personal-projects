### Removing unnecessary text from 'Salary Estimate' to create new features out of it

```python
removing_employer_provided_salary_text = dataset['Salary Estimate'].apply(lambda x: x.lower().replace("employer provided salary:", ''))
removing_employer_est_text = removing_employer_provided_salary_text.apply(lambda x: x.lower().replace("(employer est.)", '').replace("(glassdoor est.)", ''))
removing_dollar_sign = removing_employer_est_text.apply(lambda x: x.replace("$", ''))
removing_k_text = removing_dollar_sign.apply(lambda x: x.lower().replace("k", ''))
```

```python
removing_k_text.unique()
```

### Differentiating 'per hour' salary and 'fixed' salary

```python
per_hour_salary_jobs = removing_k_text.apply(lambda x: 1 if ('per hour' in x.lower()) else 0)
fixed_salary_jobs = removing_k_text.apply(lambda x: 1 if ('per hour' not in x.lower()) else 0)
```

```python
dataset = dataset.assign(parsed_salary = removing_k_text)
dataset = dataset.assign(per_hour_salary_jobs = per_hour_salary_jobs)
dataset = dataset.assign(fixed_salary_jobs = fixed_salary_jobs)
```

```python
dataset[['parsed_salary', 'per_hour_salary_jobs', 'fixed_salary_jobs']].value_counts()
```

### Calculating an Annual Salary from an Hourly Wage
Multiply the number of hours you work per week by your hourly wage. Multiply that number by 52 (the number of weeks in a year).

Example: If you make 20 dollars an hour and work 37.5 hours per week, your annual salary is 20 x 37.5 x 52 or 39,000 dollars per year.

### Creating minimum_salary	and annual_minimum_salary features

```python
min_salary_column = dataset[['parsed_salary', 'per_hour_salary_jobs']].apply(lambda x: x.parsed_salary.split(' ')[0] if (x.per_hour_salary_jobs == 1) else x.parsed_salary.split(' ')[0], axis=1)
dataset = dataset.assign(minimum_salary = min_salary_column)

working_hours = 40 # 40 hours per week
no_of_weeks = 52 # 52 weeks in a year

min_salary_column_per_year = dataset[['per_hour_salary_jobs', 'minimum_salary']].apply(lambda x: int(x.minimum_salary)*working_hours*no_of_weeks if(x.per_hour_salary_jobs == 1) else int(x.minimum_salary)*1000, axis=1)
dataset = dataset.assign(annual_minimum_salary = min_salary_column_per_year)
```

### For creating maximum_salary and annual_maximum_salary

```python
# For fixed salary jobs -> Ex: 60 (fixed salary)
max_salary_column = dataset[['parsed_salary', 'per_hour_salary_jobs']].apply(lambda x: x.parsed_salary.split(' ')[0] if( (x.per_hour_salary_jobs == 0) & (len(x.parsed_salary.split(' ')) == 1 ) ) else x.parsed_salary   , axis = 1)
dataset = dataset.assign(maximum_salary = max_salary_column)

# For fixed salary jobs -> Ex: 91-145 (fixed salary)
max_salary_column = dataset[['maximum_salary', 'per_hour_salary_jobs']].apply(lambda x: x.maximum_salary.split(' ')[2] if( (x.per_hour_salary_jobs == 0) & ('-' in x.maximum_salary) ) else x.maximum_salary   , axis = 1)
dataset = dataset.assign(maximum_salary = max_salary_column)

# For per hour jobs -> Ex: 96 per hour
max_salary_column = dataset[['maximum_salary', 'per_hour_salary_jobs']].apply(lambda x: x.maximum_salary.split(' ')[0] if( (x.per_hour_salary_jobs == 1) & (len(x.maximum_salary.split(' ')) == 3 ) ) else x.maximum_salary, axis = 1)
dataset = dataset.assign(maximum_salary = max_salary_column)

# For per hour salary jobs -> Ex: 53 - 82 per hour
max_salary_column = dataset[['maximum_salary', 'per_hour_salary_jobs']].apply(lambda x: x.maximum_salary.split(' ')[2] if( (x.per_hour_salary_jobs == 1) & ('-' in x.maximum_salary) ) else x.maximum_salary   , axis = 1)
dataset = dataset.assign(maximum_salary = max_salary_column)
```

```python
max_salary_column_per_year = dataset[['per_hour_salary_jobs', 'maximum_salary']].apply(lambda x: int(x.maximum_salary)*working_hours*no_of_weeks if(x.per_hour_salary_jobs == 1) else (int(x.maximum_salary)*1000), axis=1)
dataset = dataset.assign(annual_maximum_salary = max_salary_column_per_year)
```

## Introducing state feature from location feature

```python
import us
dataset['Location'].unique()
```

```python
dataset['Location'].value_counts()
```

```python
def findStates(dataset_feature):
    parsed_loaction_array = dataset_feature.apply(lambda x: x.split(',')[-1].strip() if(',' in x) else x )

    for i in parsed_loaction_array.index:
        try:
            parsed_loaction_array[i] = us.states.lookup(parsed_loaction_array[i]).abbr

        except:
            if parsed_loaction_array[i].lower() == 'remote':
                parsed_loaction_array[i] = "Remote"
            else:
                parsed_loaction_array[i] = str(np.nan)
    
    return parsed_loaction_array
```

```python
states_column = findStates(dataset['Location'])
dataset = dataset.assign(states = states_column)
```
