```python
import pandas as pd
import matplotlib.pyplot as plt

from feature_engine.selection import SelectBySingleFeaturePerformance
from sklearn.ensemble import RandomForestClassifier
```

```python
survey_data = pd.read_csv('data_cleaned.csv')
print(survey_data.shape)

survey_data.head()
```

# Splitting the train and test set

```python
X_train = pd.read_csv('X_train.csv')
X_test = pd.read_csv('X_test.csv')
```

```python
y_train = pd.read_csv('y_train.csv')
y_test = pd.read_csv('y_test.csv')
```

# Feature Selection

```python
rf = RandomForestClassifier(
    n_estimators = 6, random_state = 12, n_jobs=4)

sel = SelectBySingleFeaturePerformance(
    variables=None,
    estimator=rf,
    scoring="accuracy",
    cv=10)

sel.fit(X_train, y_train.values.ravel())

X_train = sel.transform(X_train)
X_test = sel.transform(X_test)
```

```python
# Saving X_train and X_test
X_train.to_csv("X_train_model_features.csv", index = False)
X_test.to_csv("X_test_model_features.csv", index = False)
```
